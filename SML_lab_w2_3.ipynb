```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=2)  # reduced display precision on numpy arrays
```

```python
# np.arange(0, 20, 1)生成了一个从0到19的整数序列，步长为1。
# 这个序列包含了从0到19的所有整数，但不包括20
x = np.arange(0, 20, 1)
y = 1 + x**2
X = x.reshape(-1, 1)
# 这里的参数(-1, 1)表示将数组x重塑为一个列向量
# 其中的-1表示根据数组的大小自动计算行数。参数 -1 在这里的作用是根据数组的大小自动计算行数。
# 通过将行数设置为 -1，NumPy 会根据数组的大小和列数来自动计算行数，以确保数组的元素数量保持不变。
# 1 表示 1列。
```

```python
model_w = np.array([18.7])
model_b = -52.08341025448668
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("no feature engineering")
plt.plot(x,X@model_w + model_b, label="Predicted Value")
#X@model_w表示将矩阵X与权重向量model_w进行矩阵乘法运算
plt.xlabel("X"); plt.ylabel("y"); plt.legend(); plt.show()
```

```python
X = x**2
X = X.reshape(-1,1)
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("no feature engineering")
plt.plot(x,np.dot(X,model_w) + model_b, label="Predicted Value")
#X@model_w表示将矩阵X与权重向量model_w进行矩阵乘法运算
plt.xlabel("X"); plt.ylabel("y"); plt.legend(); plt.show()
```

```python
X = np.c_[x, x**2, x**3]  
#np.c_函数的作用是将两个或多个数组按列方向连接起来，生成一个新的二维数组。
#连接的规则是，将数组按列方向排列，即将第一个数组的列放在前面，然后是第二个数组的列，以此类推。
# 连接后的数组的行数与原始数组的行数相同。
# a = np.array([1, 2, 3])
# b = np.array([4, 5, 6])
# c = np.c_[a, b]
# [[1 4]
# [2 5]
# [3 6]]
```

```python
model_w = np.array([0.08, 0.54, 0.03])
model_b = 0.010561850529563571
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value"); plt.title("x, x**2, x**3 features")
plt.plot(x, X@model_w + model_b, label="Predicted Value"); plt.xlabel("x"); plt.ylabel("y"); plt.legend(); plt.show()
```

```python
# create target data
x = np.arange(0, 20, 1)
y = x**2

# engineer features .
X = np.c_[x, x**2, x**3]   #<-- added engineered feature
X_features = ['x','x^2','x^3']
```

```python
fig,ax=plt.subplots(1, 3, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X[:,i],y)
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("y")
plt.show()
```
